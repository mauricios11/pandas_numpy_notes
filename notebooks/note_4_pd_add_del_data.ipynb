# Agregar ó eliminar datos 
En muchas ocasiones necesitamos agregar, eliminar ó separar datos
* pandas nos ofrece varias funciones para que este proceso se vuelva mucho más sencillo

* **.head()**: la usamos para mostrar los primeros 5 elementos del dataset (0,1,2,3,4)
* **.drop()**: la usamos para eliminar columnas de la salida, pero no del df. sus argumentos son
  * especificar la columna que se quiere quitar
  * especificar el eje que afecta a las columnas `axis = 1`
     * si se quisiera afectar a las filas, sería `axis = 0`
  * `inplace = True` indicar que se quiere borrar el elemento de manera definitiva  

`df.drop(0 , axis = 0)` : eliminar la fila 0 del df

* **del** df : la usamos para eliminar algún elemento
 * `del df['col']` para eliminar columnas

### agregar columnas
Para ello simplemente colocamos el `['Nombre columna']` al invocad al df, si no existe. Se crea y se coloca al final

* `df['nueva_columna'] = np.nan` : agregar una nueva columna con *nan's*

```python
import pandas as pd
import numpy as np
```

```python
path = "../data/bestsellers_with_categories.csv"
names = ['Nombre', 'Autor', 'Aprobacion', 'Reviews', 'Precio', 'Año', 'Genero']

data = pd.read_csv(path, sep= ',', header =0, names = names)
```

```python
df_books = pd.DataFrame(data)
df = df_books.copy()
df.head(3)
```

drop

```python
#eliminar fila 0 
df.drop(0, axis = 0).head(3)

# eliminar filas 0, 1 y 2 del df
df.drop([0,1,2], axis = 0)

# rango de filas (0,10)
df.drop(range(0,10), axis = 0).head()
```

```python
# eliminar una columna
df.drop('Genero', axis = 1).head(3)
```

```python
#mostrar el número de filas 
df.shape[0]

#mostrar el número de columnas
df.shape[1]
```

Agregar valores a una nueva columna

```python
#agregar una columna que contena valores NaN
df['Nueva_columna'] = np.nan

# crear una nueva columna y agregar os valores almacenados en el array
data = np.arange(0, df_books.shape[0])
```

con `data` obtenemos una lista de 0 a 550
* 550 es el número de filas que hay en el df, al colocar `df.shape[0]` lo que obtenemos es el número de filas de ese df

```python
a = data.shape
b = data.dtype

print(f' shape de  "data": {a}  tipo: {b}')
```

```python
# con los elementos de "data" creamos una nueva columna
df["Rango"] = data
df.head(3)
```

### agregar filas a un df
usamos la función `pd.concat()` de pandas para agregar filas teniendo como argumentos:
 * en una lista colocamos:
    * el df al que queremos agregar la fila
    * el df de la nueva fila

```python

#df = df.append(df_fila , ignore_index = True)
nueva_fila = ['nombre del libro', 'autor del libro', 5.0, 10000, 100, 2023, 'Non Fiction', np.nan, 550]
columnas = df.columns
dict_fila = dict(zip(columnas, nueva_fila)) # hacemos un diccionario con la nueva fila
df_fila = pd.DataFrame([dict_fila]) #convertimos en df la nueva_fila
df = pd.concat([df,df_fila], ignore_index = True)
df
```
