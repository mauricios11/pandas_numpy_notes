# filtrando / buscando datos con .loc y .iloc

Para filtrar datos podemos hacerlo por slices y por labels
filtrando por slices:
* `df_books[:1]` *muestra las primeras dos filas*
* `df_books[::-1]` *muestra la última fila*

filtrando por etiquetas
* `df_books['Autor']` *muestra la columna de Autor de todas las filas*
*  `df_books [['Autor','Reviews']]` *muestra las columnas: Autor y reviews de todas las filas*
     * colocamos 2 dimensiones porque estamos invocando a más de una etiqueta
     
filtrando por slices y etiquetas no es posible:
* `df_books[ [ : 5], ['Nombre', 'Autor'] `]  R: Error


Cuando queremos navegar por un df `loc` y `iloc` permiten filtrar datos de manera más precisa:


**NOTA**:
* en las columnas el índice 0 son los números de fila, declarándola así `[filas, 0:]`
* en las filas el índice 0 son los headers, declarándola así: `[0:, columnas]`
* si colocamos los números sin slice `:` estamos indiando el número de columna

### .loc
Filtra según un label.
* para especificar **filas**, es común usar slices
    E/M: 
    * `df_books.loc[:1]` *muestra las primeras dos filas (contando el header)*
    * `df_books.loc[::-1]` *muestra la última fila*

* si es por **columnas** especificamos la columna que deseamos, si es más de una, debemos colocarla: `[[col1, col2]]`

* **por medio filas y columnas**: combinamos las dos anteriores
  
   E/M:
   * `df_books.loc[ [ : 5], ['Nombre', 'Autor'] ]` 
     * *muestra las primeras 5 filas de las columnas: nombre y autor*

### .iloc
filtrar según los números de índice
* es casi igual que `loc`, con la diferencia que para poder acceder a las esiquetas, debe ser por su posición en el índice y no colocando las etiquetas en cuestión

* *para filtrar con iloc especificamos las filas y columnas que necesitamos por medio de su posición en índice y colocándolas en slices*

 `df.iloc[filas, columnas]`
 
así:
`df_books.iloc[ [ : 5], [0:2] `]
  * *muestra las primeras 5 filas de las primeras dos columnas, que son: nombre y autor*

  diferencias al especificar columnas y filas
  *  `df_books.iloc[1:6, 4]` *si colocamos la declaración de esta forma, indicamos la columna numero 4, y se verá tipo Series (vector)*
  *  `df_books.iloc[1:6, :4]` *si colocamos la declaración de esta forma,indicamos la columnas con el índice 4(sin contar el límite),se verá como una matriz (DF)*

   Tanto en `loc` como en `iloc`. Cuando estamos analizando una columna con valores numéricos, es posible operarlos a la hora de hacer el filtrado
  así :
  `df_books.loc[0:5, ['Reviews']] * -1 `


  ### diferencias entre loc y iloc
  En cuanto a la diferencia entre `iloc' y `loc`
   * `iloc` utiliza índices enteros para seleccionar filas y columnas
   * `loc`  utiliza etiquetas. 
   
   Por lo tanto, cuando se usan índices enteros para seleccionar filas y columnas en un DataFrame, se debe usar iloc.

```python
import pandas as pd
```

```python
path = "../data/bestsellers_with_categories.csv"
names = ['Nombre', 'Autor', 'Aprobacion', 'Reviews', 'Precio', 'Año', 'Genero']

data = pd.read_csv(path, sep= ',', header = 0 , names= names )
```

```python
df_books = pd.DataFrame(data)
df_books
```

```python
# primeros tres libros
df_books[:3]
```

```python
#Todas las filas de las columnas: nombre y autor
df_books[['Nombre', 'Autor']]
```

Usando `loc`

```python
# filtrar las primeras 4 filas (sin contar el límite) de las columnas nombre y autor
df_books.loc[:4,['Nombre', 'Autor']]
```

```python
# mostrar las filas, desde la úlitma hasta la primera, saltando de 4 en 4 de las columnas nombre y autor
df_books.loc[:: -4,['Nombre', 'Autor']].head()
```

usando `iloc` para filtrar con iloc especificamos las filas y columnas que necesitamos por medio de suposición en índice y colocándolas en slices
`df.iloc[filas, columnas]`


```python
# mostrar las primeras 4 filas de las primeras 2 columnas (0,1,2) sin contar el límite
df_books.iloc[0:5,0:2]
```

```python
#mostrar filas de la 2  a la 5 de las primeras 4 columnas (1-4) sin contar el límite
df_books.iloc[0:2, :4]
```

```python
# filtrar los datos de las dos primeras filas de "reviews = 3"
df_books.iloc[0:2, :3]
```

```python
#no es lo mismo esto
a = df_books.iloc[1,1]

# que esto
b = df_books.iloc[1:, 1:]

# y que esto 
c = df_books.iloc[:1,:1]

print(f' iloc[1,1]: {a}\n * * *\n',
      f'"iloc[1:, 1:]": \n\n{b}',
      f' \n * * *\n "iloc[:1, :1]" :\n {c}')
```

en el caso de `a` estamos accediendo al valor ubicado en la segunda columna, segunda fila. `[1,1]` .
* pero entonces, porqué este no es el primer elemento?
  * se entiende que se está invocando al elemento "1,1", dado que el primer elemento estaría en `[0,0]`
    * que sería el elemento de la primer fila, primer columna

en el caso de `b` *desde la segunda fila, segunda columna en adelante*
 * estamos accediendo a todo el dataset desde la segunda columna, dado la primera estaría en ubicación de índice "0": `[1:, 1:]`

en el caso de `c` estamos accediendo al primer elemento de la primer fila, primer columna, porques está especificando una sola celda.
 * con `[:2,:2]` estaríamos accediendo a las dos primeras filas de las dos primeras columnas

### diferencia entre [2:, :4] y [2:, :4]

En `[0:2, 0:4]` estamos accediendo a las primeras 2 filas de las primeras 4 columnas (sin contar el límite = 4)
* es decir, estamos seleccionando todos los atos que se encuentran en filas 0 y 1 y en columnas 0, 2, 3

En `[2:, 0:4]` estamos seleccionando todas las filas a partir de la tercera y las cuatro primeras columnas
 * estamos accediendo a todos los datos que se encuentran entre las filas 2 en adelante, de las primeras 4 columnas (sin contar el límite = 4)

con iloc, seleccionamos los datos  en función de su posición de manera exclusiva. 
* Es decir, si usamos `[:2,:4]` estamos seleccionando hasta la fila 1 (sin contar el límite = 2) y hasta la columna 3 ( sin contar el límite = 4)
  * en cambio con `[2:, :4]` estamos seleccionando desde la fila 2 hasta el final, de las columnas 0,1,2,3 

 

```python
df_books.iloc[:2,4:]
```

```python
df_books.iloc[0,0]
```

calculos matemáticos con columnas numéricas

```python
#con loc
mult_loc =  df_books.loc[:5, ['Reviews', 'Precio']] * -1
# con iloc
mult_iloc = df_books.iloc[:5, 4:6] * 10

print(f' calculos en columnas numéricas con loc:\n {mult_loc}\n\n'
      f' calculos en columnas numéricas con iloc:\n {mult_iloc}')
```

notemos que en iloc se toma en cuenta el número de índice

```python
# con loc buscar todas las filas que tengan el nombre "JJ Smith"
df_books.loc[:, ['Autor']] == "JJ Smith"
```
